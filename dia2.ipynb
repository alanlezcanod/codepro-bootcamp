Recorrer los elementos de una lista de marcas e imprimir cada marca solamente si tiene 6 letras o m치s.

```python
#Hago mi lista de marcas
lista_de_marcas = ['nike', 'adidas', 'puma', 'new ballance', 'reebok', 'topper', 'kappa']

#Aca hago que "marcas" sea una variante temporal que saque marca por marca
for marcas in lista_de_marcas:
    #Aca le digo a Python con "len" que cuente los caracteres de cada palabra y que seleccione las marcas dejando las que tienen 6 o mas caracteres
    if len(marcas) >= 6:
        print(marcas)
```

Solicitar al usuario que ingrese una frase.
